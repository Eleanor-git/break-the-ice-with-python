# Question 47

### **Question**

> **_Define a class named Circle which can be constructed by a radius. The Circle class has a method which can compute the area._**

---

### Hints

> **_Use def methodName(self) to define a method._**

---



**Solutions:**

```python
import math

class Circle:

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return (self.radius ** 2) * math.pi


obj_circle = Circle(10)
obj_circle.area()
```

# Question 48

### **Question**

> **_Define a class named Rectangle which can be constructed by a length and width. The Rectangle class has a method which can compute the area._**

---

### Hints

> **_Use def methodName(self) to define a method._**

---



**Solutions:**

```python
class Rectangle:

    def __init__(self, length, width):
        self.len = length
        self.wid = width

    def area(self):
        return self.len * self.wid


obj_rec1 = Rectangle(2,4)
obj_rec1.area()
```

# Question 49

### **Question**

> **_Define a class named Shape and its subclass Square. The Square class has an init function which takes a length as argument. Both classes have a area function which can print the area of the shape where Shape's area is 0 by default._**

---

### Hints

> **_To override a method in super class, we can define a method with the same name in the super class._**

---



**Solutions:**

```python
class Shape:
    def __init__(self):
        pass
            
    def area(self):
        return 0


class Square(Shape):   
    def __init__(self, length = 0):
        super().__init__()
        self.len = length
    
    def area(self):
        return self.len ** 2

obj_square = Square(5)
print(obj_square.area())

print(Shape().area()) # 呼叫class Shape 的 area method
```

`class Square(Shape):
    def __init__(self, length=0):
        Shape.__init__(self) # <---- 這段是在呼叫Shape的contructor method，也就是parent method的contructor
        self.length = length`
        
* 	Square 繼承 Shape
* 	當你建立 Square() 物件時，Square.__init__ 會被呼叫
* 	裡面呼叫了 Shape.__init__(self) → 就是把父類別的 __init__ 也執行一次
*	這樣可以確保父類別有的初始化動作也會做，不會漏掉
* `Shape.__init__(self)` 裡面的self是child class的object


#### 但是比較推薦的寫法如下:
`super().__init__()`      

* 這種寫法比直接寫 `Shape.__init__(self)` 更好，因為：
    * 這樣比較好維護，因為如果以後要改多重繼承，super() 會自動按照 MRO (Method Resolution Order) 呼叫正確的父類別，避免你手動硬寫類名。
