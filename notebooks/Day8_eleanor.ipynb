# Question 22

### **Question:**

> **_Write a program to compute the frequency of the words from the input. The output should output after sorting the key alphanumerically._**

> **_Suppose the following input is supplied to the program:_**


New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.


> **_Then, the output should be:_**

```
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1
```

---

### Hints

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**

* My solution 1:

````python
words = input("Write a sentence: ")
# words = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."

my_dict = {}

for i in words.split():
    if i not in my_dict.keys():
        my_dict[i] = words.split().count(i)

for k, v in sorted(my_dict.items(), key=lambda x: x[0]):
    print(f"{k}:{v}")
````

* My solution 2

````python
words = input("Write a sentence:")
# words = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."

my_dict = {}
for i in words.split():
    if i not in my_dict.keys():
        my_dict[i] = 1
    else:
        my_dict[i] += 1

for k,v in sorted(my_dict.items(), key=lambda x: x[0]):
    print(f"{k}: {v}")
````

### Learnt from others' solutions
1. 先用set()讓list裡面只剩下unique items -> 再用sorted()排序 -> 計算字數

````python
words = input("Write a sentence:").split()
# words = "What do you do 2 do you do 3 do you do believe the world will become better 22 1 3".split()

for i in sorted(set(words)):
    print(f"{i}:{words.count(i)}")
````

2. 先把句子、字數裝進去dictionary裡面 -> 再排序且印出來

````python
words = input("Input a sentence: ").split()
# words = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.".split()

dict = {}

for i in words:
    dict.setdefault(i, words.count(i))

for k, v in sorted(dict.items(), key=lambda x: x[0]):
    print(f"{k}:{v}")
````

# Question 24

### **Question:**

> **_Python has many built-in functions, and if you do not know how to use it, you can read document online or find some books. But Python has a built-in document function for every built-in functions._**

> **_Please write a program to print some Python built-in functions documents, such as abs(), int(), raw_input()_**

> **_And add document for your own function_**

### Hints:


The built-in document method is __doc__


---



**Solutions:**

````python
def power(n, p):
    '''
    param n: This is any integer number
    param p: This is power over n
    return: n^p (n ot the power of p)
    '''
    return n ** p

print(power.__doc__)
print("---- built in ----")
# print(str.__doc__)
print(abs.__doc__)
````

# Question 25

### **Question:**

> **_Define a class, which have a class parameter and have a same instance parameter._**

---

### Hints:


Define an instance parameter, need add it in __init__ method.You can init an object with construct parameter or set the value later


---



**Solutions:**

````python
class Sapient:
    property = "human" # the same instance parameter

    def __init__(self, name=None, poverty=False):
        self.name = name
        self.poverty = poverty

print(Sapient.property)
# print(Sapient.get_property())
print()

# initiate an object
betty = Sapient("Betty", True)
print(betty.name)
print(betty.poverty)
print(betty.property)

print()
julia = Sapient("Julia", False)
print(julia.name)
print(julia.poverty)
print(julia.property)
````
