# Question 4

### **Question:**

> **_Write a program which accepts a sequence of comma-separated numbers from console and generate a list and a tuple which contains every number.Suppose the following input is supplied to the program:_**


34,67,55,33,12,98


> **_Then, the output should be:_**


['34', '67', '55', '33', '12', '98']
('34', '67', '55', '33', '12', '98')


### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input.
tuple() method can convert list to tuple_**

---



**Solutions:**

```python
# Sol 1:

# seq_num = input("Prompt a sequence of comma-separated numbers: ")
seq_num = "10,2,300,400,1000" # For debuging

str_idx = 0
num_lst = []
tmp_str = ''
while str_idx < len(seq_num):
    if seq_num[str_idx] == ',':
        num_lst.append(tmp_str)
        tmp_str = ''
        str_idx += 1
        continue
    else:
        tmp_str += seq_num[str_idx]
#         print(tmp_str)  # For debuging
        str_idx += 1

num_lst.append(tmp_str) # appending the last number
print(num_lst,tuple(num_lst))
```

```python
# Sol 2:
# seq_num = input("Prompt a sequence of comma-separated numbers: ")
seq_num = "10,2,300,400,1000"


seq_list = []
tmp_num = ''

for i in seq_num:
    if i != ',':
        tmp_num += i
    else:
        seq_list.append(tmp_num)
        tmp_num = ''
seq_list.append(tmp_num)  # append the "last number"
print(seq_list, tuple(seq_list))
```

```python
# seq_num = input("Prompt a sequence of comma-separated numbers: ")
seq_num = "10,2,300,400,1000"
seq_lst = seq_num.split(',')
print(seq_lst, tuple(seq_lst))
```

# Question 6

### **Question:**

> **_Write a program that calculates and prints the value according to the given formula:_**

> **_Q = Square root of [(2 _ C _ D)/H]_**

> **_Following are the fixed values of C and H:_**

> **_C is 50. H is 30._**

> **_D is the variable whose values should be input to your program in a comma-separated sequence.For example
> Let us assume the following comma separated input sequence is given to the program:_**


100,150,180


> **_The output of the program should be:_**


18,22,24


---

### Hints:

> **_If the output received is in decimal form, it should be rounded off to its nearest value (for example, if the output received is 26.0, it should be printed as 26).In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**

```python
#Sol 1 - a simple function

import math


def cal(*args):
    c = 50
    h = 30
    d_lst = list(args)
    d_lst = d_lst[0].split(',')

    ans = []
    for d in d_lst:
        d = int(d)
        ans.append(round(math.sqrt(2 * c * d / h)))

    return ans


serial_str = input("Prompt serial number: ")
print(cal(serial_str))
```

```python
# Sol 2 - capsulate within a class, but print the answer directly.

import math


class Calculator:
    c = 50
    h = 30

    def __init__(self):
        pass

    def get_d(self):
        self.str_d = input("Prompt a serial numbers, and separate them with coma: ")
        self.d_lst = self.str_d.split(',')

    def cal(self):
        for d in self.d_lst:
            d = int(d)
            ans = round(math.sqrt(2 * Calculator.c * d / Calculator.h))
            print(ans, end=',')
        print('',end='\b')


obj_cal = Calculator()
obj_cal.get_d()
obj_cal.cal()
```

```python
#Sol3 - 延伸Solution2 把 "印出答案" 這個動作包在method裡面

import math


class Calculator:
    c = 50
    h = 30

    def __init__(self):
        pass

    def get_d(self):
        self.str_d = input("Prompt a serial numbers, and separate them with coma: ")
        self.d_lst = self.str_d.split(',')

    def cal(self):
        self.ans_lst = []
        for d in self.d_lst:
            d = int(d)
            self.ans_lst.append(round(math.sqrt(2 * Calculator.c * d / Calculator.h)))

    def print_ans(self):
        for i in self.ans_lst:
            print(i, end=',')
        print('', end = '\b')


obj_cal = Calculator()
obj_cal.get_d()
obj_cal.cal()
obj_cal.print_ans()
```

#### 其實我覺得這題不適合用class，但我只是純粹想練習寫一個class
* 比較好的做法，是把公式用function包起來

```python
from math import sqrt

def cal(d):
    c, h = 50, 30
    return round(sqrt(2 * c * d / h))

d_lst = [int(d) for d in input("Prompt a serial numbers, and split them with comma: ").split(',')]

ans_lst = list(map(cal, d_lst))

for _ in ans_lst:
    print(_, end=",")
print("", end="\b")
```

```python
#把cal(d) 用 lambda function f_cal代替

from math import sqrt

f_cal = lambda d: round(sqrt(2 * 50 * d / 30)) #f_cal represents "function calculator"
d_lst = [int(d) for d in input("Prompt a serial numbers, and split them with comma: ").split(',')]
ans_lst = list(map(f_cal, d_lst))
for _ in ans_lst:
    print(_, end=",")
print("", end="\b")
```

```python
# 把lambda function直接帶入 (Namely, 不 assign f_cal)

from math import sqrt

d_lst = [int(d) for d in input("Prompt a serial numbers, and split them with comma: ").split(',')]
ans_lst = list(map(lambda d: round(sqrt(2 * 50 * d / 30)), d_lst))
for _ in ans_lst:
    print(_, end=",")
print("", end="\b")
```

# Question 7

### **Question:**

> **_Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The element value in the i-th row and j-th column of the array should be i _ j.\***

> **_Note: i=0,1.., X-1; j=0,1,¡­Y-1. Suppose the following inputs are given to the program: 3,5_**

> **_Then, the output of the program should be:_**


[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]]


---

### Hints:

> **_Note: In case of input data being supplied to the question, it should be assumed to be a console input in a comma-separated form._**

---



**Solutions:**

```python
def list_generator_2d(dim):
    x, y = int(dim.split(',')[0]), int(dim.split(',')[1]) #可以用這個方法: x, y = map(int, dim.split(','))
    result_lst = []
    for i in range(0, x):
        y_lst = []
        for j in range(0, y):
            y_lst.append(j * i)
        result_lst.append(y_lst)

    return result_lst


get_dim = input("Prompt the 2D dimension(e.g. 3,5): ")
print(list_generator_2d(get_dim))
```

# Question 9

### **Question:**

> **_Write a program that accepts sequence of lines as input and prints the lines after making all characters in the sentence capitalized._**

> **_Suppose the following input is supplied to the program:_**


Hello world

Practice makes perfect


> **_Then, the output should be:_**


HELLO WORLD

PRACTICE MAKES PERFECT


---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**

```python
# Sol 1 - quit by typing (q)

lines_lst = []

line = input("type something>> ")

while line != 'q':
    lines_lst.append(line)
    line = input("type something>> ")

for l in lines_lst:
    print(l.upper())
```

```python
# Sol 2 - quit by "control + D" on mac book terminal, but "cmd + D" in Pycharm IDE

lst_lines = []
while True:
    try:
        line = input("Type something>> ")
        lst_lines.append(line)
    except EOFError as e:  # It's "control + D" on mac book terminal, but "cmd + D" in Pycharm IDE
        # print(f"{e}\n")
        break

for l in lst_lines:
    print(l.upper())
```
