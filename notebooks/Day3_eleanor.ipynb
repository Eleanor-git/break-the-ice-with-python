# Question 10

### **Question**

> **_Write a program that accepts a sequence of whitespace separated words as input and prints the words after removing all duplicate words and sorting them alphanumerically._**

> **_Suppose the following input is supplied to the program:_**

hello world and practice makes perfect and hello world again

> **_Then, the output should be:_**

again and hello makes perfect practice world

---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input.We use set container to remove duplicated data automatically and then use sorted() to sort the data._**

---

**Solutions:**

```python
# Solution1
#用每一個字的第一個字母下去排序

words = input("Prompt a serial whitespace separated words: ")
# words = "hello world and practice makes perfect and hello world again"

lst_words = words.split(' ')
set_words = set(lst_words)
sorted_set_words = sorted(set_words, key=lambda w: w[0])  # key: sorted by the first letter
print(sorted_set_words)
```

##### Note:

* **sorted(set_words, key=lambda w: w[0])**

    * sorted這個function會iterate **set_words** 然後把items 一個一個放到lambda function的參數w裡面去，也就是像下面這段code在做的事情一樣。
    * 然後再用lambda function return的value下去排序

```python
# 測試上面那個 key=lambda w: w[0] 的lambda function
print(set_words)

f = lambda w: w[0]
for l in set_words:
    print(f(l))
```

```python
words = input("Prompt a serial whitespace separated words: ")
# words = "hello world and practice makes perfect and hello world again"

lst_words = words.split(' ')
set_words = set(lst_words)
sorted_set_words = sorted(set_words)

print(' '.join(sorted_set_words))
```

# Question 11

### **Question**

> **_Write a program which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 5 or not. The numbers that are divisible by 5 are to be printed in a comma separated sequence._**

> **_Example:_**

0100,0011,1010,1001

> **_Then the output should be:_**

1010

> **_Notes: Assume the data is input by console._**

---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

**Solutions:**

```python
# Solution 1 - 用自己寫的split()

def split_by_comma(binnum_seq):
    lst = []
    num = ''
    for i in binnum_seq:
        if i != ',':
            num += i
        else:
            lst.append(num)
            num = ''
    lst.append(num)
    return lst


bin_num = input("Prompt a sequence of comma separated 4 digit binary numbers>> ")
# bin_num = '0100,0011,1010,1001,1100100,1111100011,111000010'
split_lst = split_by_comma(bin_num)

for i in split_lst:
    num = int(i, base=2)  # trans binary into decimal number
    if num % 5 == 0:
        print(bin(num)[2:], end=',')
print('', end='\b')
```

```python
#sol - 用python built-in的split()

bin_num = input("Prompt a sequence of comma separated binary numbers: ")
# bin_num = '0100,0011,1010,1001,1100100,1111100011,111000010'

for i in bin_num.split(','):
    if int(i, base=2) % 5 == 0:
        print(bin(int(i, base=2))[2:], end=',')
print('', end='\b')
```

# Question 12

### **Question:**

> **_Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number.The numbers obtained should be printed in a comma-separated sequence on a single line._**

---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

**Solutions:**

```python
num_seq = input("Prompt comma separated upperbound and lowerbound numbers: ")
boundry = list(map(int, num_seq.split(',')))

for i in range(boundry[0], boundry[1] + 1):
    if i % 2 == 0:
        print(i, end=',')
print('', end='\b')
```

# Question 13

### **Question:**

> **_Write a program that accepts a sentence and calculate the number of letters and digits._**

> **_Suppose the following input is supplied to the program:_**

hello world! 123

> **_Then, the output should be:_**

LETTERS 10

DIGITS 3

---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

**Solutions:**

```python
import re


def count(lst):
    count = 0
    for item in lst:
        for letter in item:
            count += 1
    return count


seq = input("Enter the a sentence: ")
num = re.findall(r'\d+', seq)
letter = re.findall(r'[a-z,A-Z]+', seq)

print(f"{num} has {count(num)} digits in total.")
print(f"{letter} has {count(letter)} letters in total.")
```

```python
## 別的solution

word = input("Prompt a sentence: ")

digits, letters = 0, 0
for i in word:
    if i.isalpha():
        letters += 1
    elif i.isdigit():
        digits += 1
        
print(f"#Digits: {digits}")
print(f"#Letters: {letters}")
```
