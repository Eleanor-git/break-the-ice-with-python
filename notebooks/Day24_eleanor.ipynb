# Question 101

### **Question**

> **_You are given a string.Your task is to count the frequency of letters of the string and print the letters in descending order of frequency._**

> **_If the following string is given as input to the program:_**
>
> aabbbccde
>
> **_Then, the output of the program should be:_**
>
> b 3
> a 2
> c 2
> d 1
> e 1

### Hints

> **_Count frequency with dictionary and sort by Value from dictionary Items_**

---

**Solutions:**

```python
given_str = input("Input a string: ")
# given_str = "aabbbccde"

my_dict = {}
for _ in given_str:
    if _ not in my_dict:
        my_dict[_] = 1
    else:
        my_dict[_] += 1

my_lst = sorted(my_dict.items(), key=lambda x: x[1], reverse=True)

for _ in my_lst:
    print(*_, end = ' ')
```

```python
# given_str = input("Input a string: ")
given_str = "aabbbccde"
given_str_set = set(given_str)

my_dict = {}

for _ in given_str_set:
    my_dict[_] = given_str.count(_)

for _ in sorted(my_dict.items(), key=lambda x: x[1], reverse=True):
    print(*_, end=" ")
```

```python
# given_str = input("Input a string: ")
given_str = "aabbbccde"
str_cnt_dict = {_ : given_str.count(_) for _ in set(given_str)}

for _ in sorted(str_cnt_dict.items(), key = lambda x: x[1], reverse = True):
    print(*_, end = ' ')
```

# Question 102

### **Question**

> **_Write a Python program that accepts a string and calculate the number of digits and letters._**

**Input**

> Hello321Bye360

**Output**

> Digit - 6

> Letter - 8

---

### Hints

> **_Use isdigit() and isalpha() function_**

---

**Solution:**

```python
# given_str = input("Enter a string: ")
given_str = "Hello321Bye360"

num_digits = 0
num_letters = 0

for item in given_str:
    if item.isdigit():
        num_digits += 1
    elif item.isalpha():
        num_letters += 1

print("Digit - {}".format(num_digits))
print("Letter - {}".format(num_letters))
```

```python
import re

# given_str = input("Enter a string: ")
given_str = "Hello321Bye360"

num_digits = len(re.findall(r"\d", given_str))
num_letters = len(re.findall(r"[a-z,A-Z]", given_str))

print("Digit - {}".format(num_digits))
print("Letter - {}".format(num_letters))
```

# Question 103

### **Question**

> **_Given a number N.Find Sum of 1 to N Using Recursion_**

**Input**

> 5

**Output**

> 15

---

### Hints

> **_Make a recursive function to get the sum_**

---

**Solution:**

```python
# recursive function

def sum_till_num(num):
    if num == 0:
        return 0
    if num > 0:
        return num + sum_till_num(num - 1)
    return f"Given number: \"{num}\" is not a positive integers."

n = -1
sum_till_num(n)
```

```python
#recursive function 別人的答案

def sum_till_num(num):
    if num == 0:
        return num
    return num + sum_till_num(num - 1)

n = 100
sum_till_num(n)
```
