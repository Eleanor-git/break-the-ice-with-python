# Question 14

### **Question:**

> **_Write a program that accepts a sentence and calculate the number of upper case letters and lower case letters._**

> **_Suppose the following input is supplied to the program:_**


Hello world!


> **_Then, the output should be:_**


UPPER CASE 1

LOWER CASE 9


---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**

```python
input_string = input("Prompt a random sentence: ")
# input_string = 'Hello, Today Is an anti-immigration Protest Day.'

upper_cnt, lower_cnt = 0, 0
for letter in input_string:
    if letter.isupper():
        upper_cnt += 1
    elif letter.islower():
        lower_cnt += 1
#     else: #其他符號
#         print(f"{letter} is neither upper nor lower")

print("UPPER CASE: {} ".format(upper_cnt))
print("LOWER CASE: {0}".format(lower_cnt))
```

---
### Solutions From Others

* 下面的第一個例子很特別
    * 是利用Python 布林值True = 1, False = 0的特性下去做計算
    * 所以當你寫：`lower += i.islower()`
        * 如果 `i.islower()` 是 True，就相當於 `lower += 1`
        * 如果 `i.islower()` 是 False，就相當於 `lower += 0`（數值不變）

```python
word = input()
upper, lower = 0, 0

for i in word:
    lower += i.islower()
    upper += i.isupper()

print("UPPER CASE {0}\nLOWER CASE {1}".format(upper, lower))
```

```python
word = input()
upper = sum(1 for i in word if i.isupper()) # sum function cumulatively sum up 1's if the condition is True
lower = sum(1 for i in word if i.islower())

print("UPPER CASE {0}\nLOWER CASE {1}".format(upper, lower))
```

```python
word = input()
upper, lower = 0, 0

for i in word:
    if "a" <= i and i <= "z":
        lower += 1
    if "A" <= i and i <= "Z":
        upper += 1

print("UPPER CASE {0}\nLOWER CASE {1}".format(upper, lower))
```

# Question 15

### **Question:**

> **_Write a program that computes the value of a+aa+aaa+aaaa with a given digit as the value of a._**

> **_Suppose the following input is supplied to the program:_**


9


> **_Then, the output should be:_**


11106


---
### 題目中文解釋：

計算 a + aa + aaa + aaaa
給你一個數字 a，你要算出這四個數加起來的結果

用題目舉的例子：
輸入：9

代表
*	第一項：a = 9
*	第二項：aa = 99
*	第三項：aaa = 999
*	第四項：aaaa = 9999

然後把它們加起來：

---
### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**

#### Solution 1: For loop

```python
x = input("Prompt a number: ")
# x = '9'
# 9 + 99 + 999 + 9999
# x + xx + xxx + xxxx


looping_time = 4
ans = 0
for i in range(1, looping_time + 1):
    ans += int(x * i)

print(ans)
```

#### Solution2 : while loop (Descending)

```python
x = input("Prompt a number: ")
# x = '10'

# while loop (Descending)
looping_time = 4
total = 0

while looping_time > 0:
    total += int(x * looping_time)
    looping_time -= 1

print(total)
```

#### Solution3: while loop (Ascending)

```python
x = input("Prompt a number: ")
# x = '1'

# while loop (Descending)
loop_cnt = 1
ans = 0
while loop_cnt <= 4:
    ans += int(x * loop_cnt)
    loop_cnt += 1

print(ans)
```
