# Question 1

### **Question:**

> **_Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5,
> between 2000 and 3200 (both included).The numbers obtained should be printed in a comma-separated sequence on a single line._**

---

### Hints:

> **_Consider use range(#begin, #end) method._**

---



**Solutions:**

- **Using for loops**

```python
def find_num(lowerbound, upperbound, div_num, mul_num):
    lst = []
    for _ in range(lowerbound, upperbound):
        if _ % div_num ==0 and _ % mul_num != 0:
            lst.append(_)
    return lst   

print(find_num(2000, 3201, 7, 5))
```

* 另一種偷懶的方法,用filter()

```python
def filter_condition(num):
    return num % 7 == 0 and num % 5 != 0


given_num = list(range(2000, 3201))
filtered_lst = list(filter(filter_condition, given_num))
print(filtered_lst)
```

# Question 2

### **Question:**

> **_Write a program which can compute the factorial of a given numbers.The results should be printed in a comma-separated sequence on a single line.Suppose the following input is supplied to the program: 8
> Then, the output should be:40320_**

---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---



**Solutions:**


- **Using For Loop**

```python
def factorial(given_num):
    multiple_seed = 1
    for _ in range(1, given_num + 1):
        multiple_seed *= _
    print(multiple_seed)


factorial(int(input("Enter a number: ")))
```

- **Using While Loop**

```python
given_num = int(input("Enter a number: "))

if given_num == 0:
    fact = 1
else:
    fact = given_num
    base = given_num - 1
    while base > 0:
        fact = fact * base
        base = base - 1

print(fact)
```

# Question 3

### **Question:**

> **_With a given integral number n, write a program to generate a dictionary that contains (i, i x i) such that is an integral number between 1 and n (both included). and then the program should print the dictionary.Suppose the following input is supplied to the program: 8_**

> **_Then, the output should be:_**


{1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64}


---

### Hints:

> **_In case of input data being supplied to the question, it should be assumed to be a console input.Consider use dict()_**

**Solutions:**

####  Eleanor's Solution1: 

```python
given_num = int(input("Prompt an integer number: "))
print(dict(zip(range(1, given_num + 1), map(lambda num: num * num, range(1, given_num + 1)))))
```

#### Eleanor's Solution2:
* with **For Loop**

```python
# given_num = int(input("Prompt an integer number: "))
given_num = 8

my_dict = {}
for i in range(1, given_num + 1):
    my_dict[i] = i * i
print(my_dict)
```

#### Eleanor's Solution3:
* with **While Loop descending**

```python
# given_num = int(input("Prompt an integer number: "))
given_num = 9

n = given_num
my_dict = {}
while n > 0:
    my_dict[n] = n * n
    n = n - 1
print(my_dict)
```

* with **While Loop ascending**

```python
given_num = int(input("Prompt an integer number: "))
# given_num = 9

n = 1
my_dict = {}
while n <= given_num:
    my_dict[n] = n * n
    n += 1
print(my_dict)
```
