```python
def bin_search2(lst, target):
    lower, upper = 0, len(lst) - 1
    lst.sort()

    while lower < upper:
        mid = (lower + upper) // 2

        if lst[mid] == target:
            return mid

        elif lst[mid] < target:
            lower = mid + 1
        else:
            upper = mid

    if lst[upper] == target:
        return upper
    else:
        return f"{target} doesn't exist in the list."

lst = [5, 6, 14, 22, 28, 30, 55, 60]
target = 1

print(bin_search(lst, target))
```
