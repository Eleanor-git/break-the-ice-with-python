```python
# get_str_width = input("Prompt a string and width split with a blank: ")
# s = get_str_width.split()[0]
# width = int(get_str_width.split()[1])

s = "ABCDEFGHIJKLIMNOQRSTUVWXYZ"
width = 4

for idx in range(0, len(s), width):
    print(s[idx:idx + width], end=" ")
```

```python
s = "ABCDEFGHIJKLIMNOQRSTUVWXYZ"
width = 4


def idx(length, width):
    start = 0
    end = start + width
    while end <= length:
        end = start + width
        yield start, end
        start = end


idx_generator = idx(len(s), width)

for start, end in idx_generator:
    print(s[start:end], end=" ")
```

# Question 97

### **Question**

> **_You are given an integer, N. Your task is to print an alphabet rangoli of size N. (Rangoli is a form of Indian folk art based on creation of patterns.)_**

> **_Different sizes of alphabet rangoli are shown below:_**
>
> $size = 3$
>
> ----c----

> --c-b-c--

> c-b-a-b-c

> --c-b-c--

> ----c----
>
> $size = 5$
>
> --------e--------

> ------e-d-e------

> ----e-d-c-d-e----

> --e-d-c-b-c-d-e--

> e-d-c-b-a-b-c-d-e

> --e-d-c-b-c-d-e--

> ----e-d-c-d-e----

> ------e-d-e------

> --------e--------

### Hints

> **_First print the half of the Rangoli in the given way and save each line in a list. Then print the list in reverse order to get the rest._**

---

**Solutions:**

```python
#Solution 1
my_char = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j','k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v',
               'w', 'x', 'y', 'z']


def gen_char(num):
    left_list = my_char[:num]
    while left_list:
        reversed_left_list = list(reversed(left_list))
        left_list.pop(0)

        right_list = left_list
        yield reversed_left_list, right_list


given_int = 5
letter_gen = gen_char(given_int)

downward_lst = []
for left, right in letter_gen:
    downward_lst.append(left + right)

# # --- printing uppward part ---
upward_lst = list(reversed(downward_lst))
num_dash = given_int * 2 - 2

for l in upward_lst:
    dash = '-' * num_dash
    num_dash = num_dash - 2
    print(dash + '-'.join(l) + dash)

# # --- printing downward part ---
dash = '--'
for l in downward_lst[1:]:
    print(dash + '-'.join(l) + dash)
    dash += '--'
```

```python
# Sol2:

my_char = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

num = 26
final_left_lst, final_right_lst = [], []

range_lst = my_char[:num]

while len(range_lst):
    left_lst = list(reversed(range_lst))
    final_left_lst.append(left_lst)
    range_lst.pop(0)
    right_list = range_lst.copy()
    final_right_lst.append(right_list)

# Organizing the format and printing out
# up
num_dash = (num - 1) * 2
dash = '-' * num_dash

print(dash + final_left_lst[-1][0] + dash)
for idx in range(2, len(final_right_lst)):
    num_dash = num_dash - 2
    dash = '-' * num_dash
    print(dash + '-'.join(final_left_lst[-idx] + final_right_lst[-idx]) + dash)

# down
dash = '--'
for idx in range(len(final_right_lst)):
    print(dash * idx + '-'.join(final_left_lst[idx] + final_right_lst[idx]) + dash * idx)
```

```python
import string


def print_rangoli(size):
    n = size
    alph = string.ascii_lowercase
    width = 4 * n - 3 

    ans = []
    for i in range(n):
        left = "-".join(alph[n - i - 1 : n])
        mid = left[-1:0:-1] + left # the type of mid is "str"
        final = mid.center(width, "-")
        ans.append(final)

    if len(ans) > 1:
        for i in ans[n - 2 :: -1]:
            ans.append(i)
    ans = "\n".join(ans)
    print(ans)


if __name__ == "__main__":
    n = int(input("Prompt the size of rangoli: "))
    print_rangoli(n)
```

#### `width = 4 * n - 3` 是怎麼得出來的?

> 以size = 3舉例
>* c-b-a-b-c
>* n = alphabet a b c的數量
>* 第1個(n-1): 是a-b-c中間的dash數量
>* 第2個(n-1):是左邊字母數量 i.e. c, b
>* 第3個(n-1):是左邊字母中間的dash i.e. c-b-

全部加起來並且移項: `n + (n-1) + (n-1) +(n-1) = 4n -3`

#### `left[-1:0:-1]` 是reverse,但是只取到 idx = 1
>* `'abcde'[-1:0:-1] = 'edcb'`

#### `left = "-".join(alph[n - i - 1 : n])`

```python
alpha = 'abcdefghijklmnopqrstuvwxyz'
size = 3
alpha[size -1:size] #alpha [2: 3] 因為 i = 0，所以要多減掉一個1
alpha[size -2:size] #alpha [1: 3] i = 1
alpha[size -3:size] #alpha [0: 3] i = 2

#把range範圍改成如下也可以
for idx in range(1, size + 1):
    print(f"idx: {idx}")
    print("-".join(alpha[size - idx: size]))
```

# Question 98

### **Question**

> **_You are given a date. Your task is to find what the day is on that date._**

**Input**

> **_A single line of input containing the space separated month, day and year, respectively, in MM DD YYYY format._**
>
> 08 05 2015

**Output**

> **_Output the correct day in capital letters._**
>
> WEDNESDAY

---

### Hints

> **_Use weekday function of calender module_**

---

**Solution:**

```python
import calendar


def trans_weekday(weekday):
    if weekday == 0:
        return "Monday"
    elif weekday == 1:
        return "Tuesday"
    elif weekday == 2:
        return "Wednesday"
    elif weekday == 3:
        return "Thursday"
    elif weekday == 4:
        return "Friday"
    elif weekday == 5:
        return "Saturday"
    elif weekday == 6:
        return "Sunday"
    else:
        return "Unknown"


# given_date = int(input("Prompt a date (e.g. 10 05 1991): "))
given_date = "10 05 1991"

day, month, year = given_date.split()[0], given_date.split()[1], given_date.split()[2]
weekday = calendar.weekday(int(year), int(month), int(day))

print(trans_weekday(weekday))
```

# Question 99

### **Question**

> **_Given 2 sets of integers, M and N, print their symmetric difference in ascending order. The term symmetric difference indicates those values that exist in either M or N but do not exist in both._**

**Input**

> **_The first line of input contains an integer, M.The second line contains M space-separated integers.The third line contains an integer, N.The fourth line contains N space-separated integers._**
>
> 4
> 2 4 5 9
> 4
> 2 4 11 12

**Output**

> **_Output the symmetric difference integers in ascending order, one per line._**
>
> 5
> 9
> 11
> 12

---

### Hints

> **_Use \'^\' to make symmetric difference operation._**

---

**Solution:**

```python
# Sol1:


def get_set():
    num_element = int(input("Prompt an integer number for set1: "))
    element = [num_element] + list(map(int, input(f"Prompt {num_element} space-separated integers: ").split()))
    return element


m = set(get_set())
print("----------")
n = set(get_set())
print("----------")
print("set m: {}".format(m))
print("set n: {}".format(n))
print("---------")
print("Symmetric difference: {}".format(sorted(m ^ n)))
```
