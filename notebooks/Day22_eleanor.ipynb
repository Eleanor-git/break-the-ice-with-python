# Question 90

### **Question**

> **_Please write a program which count and print the numbers of each character in a string input by console._**

> **_Example:
> If the following string is given as input to the program:_**

```
abcdefgabc
```

> **_Then, the output of the program should be:_**

```
a,2
c,2
b,2
e,1
d,1
g,1
f,1
```

### Hints

> **_Use dict to store key/value pairs.
> Use dict.get() method to lookup a key with default value._**

---

**Solutions:**

````python
#Sol1: Instead of using dict.get() , applying if condition

# my_str = input("input a string: ")
my_str = 'abcdefgabc'

my_dict = {}
for char in my_str:
    if char not in my_dict.keys():
        my_dict[char] = my_str.count(char)
print(my_dict)
````

````python
# Sol2: solving the problem by using dict.get()
# 後面蓋掉前面

# my_str = input("input a string: ")
my_str = 'abcdefgabc'

my_dict = {}
for char in my_str:
    
    my_dict[char] = my_dict.get(char, my_str.count(char))

print(my_dict)
````

````python
# Sol3: Applying "Dictionary Comprehension" way to solve.
# 後面蓋掉前面
# my_str = input("input a string: ")
my_str = 'abcdefgabc'

{char:my_str.count(char) for char in my_str}
````

# Question 92

### **Question**

> **_Please write a program which accepts a string from console and print the characters that have even indexes._**

> **_Example:
> If the following string is given as input to the program:_**
```
H1e2l3l4o5w6o7r8l9d
```
> **_Then, the output of the program should be:_**
```
Helloworld
```
### Hints

> **_Use list[::2] to iterate a list by step 2._**

---



**Solutions:**

````python
# Sol1

my_str = "H1e2l3l4o5w6o7r8l9d"
my_str[::2]
````

````python
# Sol2

my_str = "H1e2l3l4o5w6o7r8l9d"

new_str = ""
for idx in range(len(my_str)):
    if idx %2 == 0:
        new_str += my_str[idx]
new_str
````

````python
#Sol3

my_str = "H1e2l3l4o5w6o7r8l9d"

new_str = ""
for idx, value in list(enumerate(my_str)):
    if idx %2 == 0:
        new_str += value
new_str
````

# Question 94

### **Question**

> **_Write a program to solve a classic ancient Chinese puzzle:
> We count 35 heads and 94 legs among the chickens and rabbits in a farm. How many rabbits and how many chickens do we have?_**

---

### Hints

> **_Use for loop to iterate all possible solutions._**

---

**Solution:**

* The key is the following statements:

1. `#chickens + #rabbits = 35`
2. `#chickens * 2 + #rabbits * 4 = 94`

````python
head = int(input("Prompt the number of heads: "))
legs = int(input("Prompt the number of legs: "))
ans = False

for chk_num in range(head + 1):
    for rabbit_num in range(head + 1):
        if chk_num + rabbit_num == head and 2 * chk_num + 4 * rabbit_num == legs:
            ans = True
            break

    if ans:
        break

if ans:
    print(f"Number of Chickens: {chk_num}\nNumber of Rabbits: {rabbit_num}")
else:
    print("There's no answer.")
````

````python
def chicken_rabbit(heads, legs):
    rabbits = (legs - 2 * heads) / 2
    chickens = heads - rabbits
    if rabbits.is_integer() and chickens.is_integer() and rabbits >= 0 and chickens >= 0:
        return int(chickens), int(rabbits)
    else:
        return "无解"

heads = int(input("请输入头的数量: "))
legs = int(input("请输入腿的数量: "))

result = chicken_rabbit(heads, legs)

if result == "无解":
    print("输入的头和脚的数量无法满足鸡兔同笼问题")
else:
    print(f"鸡的数量为: {result[0]}")
    print(f"兔子的数量为: {result[1]}")
````

````python
#ChatGPT 窮舉法

def chicken_rabbit(heads, legs):
    solutions = []
    for chickens in range(heads + 1):
        rabbits = heads - chickens
        if 2 * chickens + 4 * rabbits == legs:
            solutions.append((chickens, rabbits))
    return solutions


print(chicken_rabbit(35, 94))  # 35頭，94腳
````

````python
# ChatGPT 代數法
def chicken_rabbit_algebra(heads, legs):
    rabbits = (legs - 2 * heads) // 2
    chickens = heads - rabbits
    if rabbits >= 0 and chickens >= 0 and 2 * chickens + 4 * rabbits == legs:
        return chickens, rabbits
    else:
        return None

print(chicken_rabbit_algebra(35, 94))
````
