# Question 80

### **Question**

> **_Please write a program to print the list after removing even numbers in [5,6,77,45,22,12,24]._**

---

### Hints

> **_Use list comprehension to delete a bunch of element from a list._**

---



**Solutions:**

```python
# My first solution

lst = [5,6,77,45,22,12,24]
[ _ for _ in lst if _ %2 != 0]
```

```python
# My second solution inspired by others' solution

lst = [5,6,77,45,22,12,24]

list(filter(lambda x : x % 2 != 0, lst))
```

# Question 82

### **Question**

> **_By using list comprehension, please write a program to print the list after removing the 0th, 2nd, 4th,6th numbers in [12,24,35,70,88,120,155]._**

---

### Hints

> **_Use list comprehension to delete a bunch of element from a list.
> Use enumerate() to get (index, value) tuple._**

---



**Solutions:**

```python
# Solution1 比較方便且符合題目使用comprehension的寫法

lst = [12,24,35,70,88,120,155]
[v for k,v in tuple(enumerate(lst)) if k %2 != 0]



# # Solution1 without comprehension
# tup = tuple(enumerate(lst))
# odd_lst = []

# for k,v in tup:
#     if k %2 != 0:
#         odd_lst.append(v)
# odd_lst
```

```python
# Solution2 - without using enumerate()
lst = [12, 24, 35, 70, 88, 120, 155]
[lst[idx] for idx in range(len(lst)) if idx %2 != 0]


# # Solution2 without comprehension
# lst = [12, 24, 35, 70, 88, 120, 155]
# odd_lst = []

# for i in range(len(lst)):
#     if i % 2 != 0:
#         odd_lst.append(lst[i])
# odd_lst
```

---

### Hints

> **_Use list comprehension to delete a bunch of element from a list.
> Use enumerate() to get (index, value) tuple._**

---



**Solutions:**

```python
# range(2,5) = 2~4

lst = [12,24,35,70,88,120,155]
print(lst[2])
print(lst)
 lst.pop(2)
lst
```

```python
#Sol1

lst = [12,24,35,70,88,120,155]
[v for k, v in tuple(enumerate(lst)) if k < 2 or k > 4]



# # Solution1 without comprehension
# lst = [12,24,35,70,88,120,155]

# new_lst = []
# for k,v in tuple(enumerate(lst)):
#     if k != 2 and k !=3 and k != 4:
#         new_lst.append(v)
# new_lst
```

```python
#Sol2

lst = [12, 24, 35, 70, 88, 120, 155]
tup = tuple(enumerate(lst))
print(list(filter(lambda x: x[0] < 2 or x[0] > 4, tup)))

[v for k, v in list(filter(lambda x: x[0] < 2 or x[0] > 4, tup))]  # 如果只想要value,不需要index的話，可以這樣做
```

```python
# Sol3 用集合的方式

lst = [12,24,35,70,88,120,155]
tup = tuple(enumerate(lst))

list(filter(lambda x: {x[0]}.difference({2,3,4})  , tup))
```

# Question 84

### **Question**

> **_By using list comprehension, please write a program generate a 3\*5\*8 3D array whose each element is 0._**

---

### Hints

> **_Use list comprehension to make an array._**

---

**Solution:**

```python
# Sol1: 

small_lst = []
mid_lst = []
final_lst1 = []

for _ in range(8):
    small_lst.append(0)

for _ in range(5):
    mid_lst.append(small_lst)

for _ in range(3):
    final_lst1.append(mid_lst)

final_lst1
```

```python
small_lst = [0 for _ in range(8)]
mid_lst = [small_lst for _ in range(5)]
final_lst2 = [mid_lst for _ in range(3)]

final_lst2
```

```python
final_lst3 = [[[0 for _ in range(8)] for _ in range(5)] for _ in range(3)]
```

---
