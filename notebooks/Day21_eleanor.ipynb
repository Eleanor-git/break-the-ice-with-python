# Question 85

### **Question**

> **_By using list comprehension, please write a program to print the list after removing the 0th,4th,5th numbers in [12,24,35,70,88,120,155]._**

---

### Hints

> **_Use list comprehension to delete a bunch of element from a list.Use enumerate() to get (index, value) tuple._**

---



**Solutions:**

```python
#Solution1 - 按照題目要求的答案
lst = [12, 24, 35, 70, 88, 120, 155]
[_[1] for _ in list(filter(lambda x: x[0] != 0 and x[0] != 4 and x[0] != 5,list(enumerate(lst))))]
```

```python
#Solution2

lst = [12, 24, 35, 70, 88, 120, 155]
#      0   1   2   3   4   5    6
new_lst = []

for idx in range(len(lst)):
    if idx != 0 and idx != 4 and idx != 5:
        new_lst.append(lst[idx])

print(new_lst)
```

> **_With two given lists [1,3,6,78,35,55] and [12,24,35,24,88,120,155], write a program to make a list whose elements are intersection of the above given lists._**

---

### Hints

> **_Use set() and "&=" to do set intersection operation._**

---



**Solutions:**

```python
# Solution 1

lst1 =  [1,3,6,78,35,55]
lst2 =  [12,24,35,24,88,120,155]

set(lst1) & set(lst2)
```

# Question 89

### **Question**

> **_Define a class Person and its two child classes: Male and Female. All classes have a method "getGender" which can print "Male" for Male class and "Female" for Female class._**

---

### Hints

> **_Use Subclass(Parentclass) to define a child class._**

---

**Solution:**

```python
class Person(object):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_instance(cls, obj):
        return cls(obj.name, obj.age), obj.getGender()


class Male(Person):
    def getGender(self):
        self.gender = "Male"
        return self.gender


class Female(Person):
    def getGender(self):
        self.gender = "Female"
        return self.gender


man = Male(name="Jerry", age=30)
print(man.name, man.age, man.getGender())

woman = Female(name="Rose", age=99)
rose, gender = Person.from_instance(woman)
print(rose.name, rose.age, gender)
print(rose)
```

```python
class Person(object):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_instance(cls, obj):
        return cls(obj.name, obj.age), obj
    #   return tuple(Person object, Femal object)


class Male(Person):
    def getGender(self):
        self.gender = "Male"
        return self.gender


class Female(Person):
    def getGender(self):
        self.gender = "Female"
        return self.gender


man = Male(name="Jerry", age=30)
print(man.name, man.age, man.getGender())

woman = Female(name="Rose", age=99)
print(Person.from_instance(woman)[0], Person.from_instance(woman)[1])
print(Person.from_instance(woman)[0].name, Person.from_instance(woman)[0].age,
      Person.from_instance(woman)[1].getGender())
```
