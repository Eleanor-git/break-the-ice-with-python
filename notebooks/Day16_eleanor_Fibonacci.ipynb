> $f(n)=f(n-1)+100$ when n>0

> **_with a given n input by console (n>0)._**

> **_Example:
> If the following n is given as input to the program:_**

> **_Please write a program to compute the value of f(n) with a given n input by console._**

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n > 1:
        return fibonacci(n - 1) + fibonacci(n - 2)
    return f"n should be and integer and >= 0"


num = int(input("Prompt a number >= 0: "))
print(fibonacci(num))
```

# Question 62

### **Question**

> **_The Fibonacci Sequence is computed based on the following formula:_**


> $f(n)=0$ if n=0

> $f(n)=1$ if n=1

> $f(n)=f(n-1)+f(n-2)$ if n>1


> **_Please write a program to compute the value of f(n) with a given n input by console._**

> **_Example:
> If the following n is given as input to the program:_**

> 7

> **_Then, the output of the program should be:_**

> 0,1,1,2,3,5,8,13

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

### Hints

> **_We can define recursive function in Python.
> Use list comprehension to generate a list from an existing list.
> Use string.join() to join a list of strings._**

---



### Solutions:

```python
n = 7
fib_list = [0] * (n + 1)


def f(n):
    if n < 2:
        fib_list[n] = n
        return fib_list[n]
    elif n >= 2:
        fib_list[n] = f(n - 1) + f(n - 2)
        return fib_list[n]
    else:
        return f"error"


f(n)
print(fib_list)
# print(",".join([str(_) for _ in fib_list]))
```

**Chat GPT給的答案 (遞迴＋generator)**

**原理**:

* fib_seq(n, a, b)：a 是當前數字，b 是下一個數字。
* 每次 yield a，然後遞迴呼叫，把 (b, a+b) 傳進去。
* n 控制還要產生幾個數字。

這樣就只需要 一個函式，同時做到遞迴 + 序列產生。

```python
def fib_seq(n, a=0, b=1):
    if n == 0:
        yield a   # 回傳 0
        return
    yield a
    yield from fib_seq(n - 1, b, a + b)
```

```python
# 遞迴和generator分開寫的版本

def fib(n):
    if n < 2:
        return n
    elif n >= 2:
        return fib(n - 1) + fib(n - 2)


def fib_gen(num):
    for n in range(num + 1):
        if n < 2:
            yield n
        elif n >= 2:
            yield fib(n - 1) + fib(n - 2)


','.join([str(i) for i in fib_gen(10)])
```

```python
# generator only 的版本(沒有遞迴)

def fib(num):
    n2 = 0
    n1 = 1

    for n in range(num + 1):
        if n < 2:
            yield n
        elif n >= 2:
            fn = n1 + n2
            yield fn

            n2 = n1
            n1 = fn


for i in fib(7):
    print(i, end=',')
print('\b')

','.join([str(i) for i in fib(10)]) # comprehension expression
```

# Question 63

### **Question**

> **_Please write a program using generator to print the even numbers between 0 and n in comma separated form while n is input by console._**

> **_Example:
> If the following n is given as input to the program:_**

> 10

> **_Then, the output of the program should be:_**

> 0,2,4,6,8,10

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

### Hints

> **_Use yield to produce the next value in generator._**

---

**Solution:**

```python
def even_num_generator(n):
    for i in range(n + 1):
        if i % 2 == 0:
            yield i


num = int(input("Prompt an integer: "))
# num = 10
obj_even_gen = even_num_generator(num)

try:
    for i in range(num + 1):
        print(next(obj_even_gen), end=',')
except StopIteration:
    print('\b')
```

#### while descending 寫法

```python
def even_num_generator(n):
    while n >= 0:
        if n % 2 == 0:
            yield n
        n -= 1

# num = int(input("Prompt an integer: "))
num = 10
obj_even_gen = even_num_generator(num)

try:
    for i in range(num + 1):
        print(next(obj_even_gen), end=',')
except StopIteration:
    print('\b')
```

#### while ascending 寫法

```python
def even_num_generator(n):
    i = 0
    while i <= n:
        if i % 2 == 0:
            yield i
        i += 1


# num = int(input("Prompt an integer: "))
num = 10
obj_even_gen = even_num_generator(num)

try:
    for i in range(num + 1):
        print(next(obj_even_gen), end=',')
except StopIteration:
    print('\b')
```

# Question 64

### **Question**

> **_Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console._**

> **_Example:
> If the following n is given as input to the program:_**

> 100

> **_Then, the output of the program should be:_**

> 0,35,70

> **_In case of input data being supplied to the question, it should be assumed to be a console input._**

---

### Hints

> **_Use yield to produce the next value in generator._**

---



**Solutions:**

```python
def gen_div_5n7(n):
    if n >= 0:
        for i in range(n + 1):
            if i % 5 == 0 and i % 7 == 0:
                yield i
        return None
    else:
        return f"Given number is Error!"


num = int(input("Prompt a number: "))
obj_gen = gen_div_5n7(num)

try:
    for _ in range(num + 1):
        print(next(obj_gen), end=',')
except StopIteration:
    print('\b')
```
